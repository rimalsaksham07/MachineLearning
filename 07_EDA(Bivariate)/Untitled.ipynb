# 1.Scatterplot (Numerical - Numerical)

```python
# if we have two different different column and we want to analyze their data 
# step-1 is to know the given data is numerical or datatype of the given data
```

```python
# both have numercial data 
```

```python
sns.scatterplot( x = tips['total_bill'], y = tips['tip']) # this is bivariate 
# now we also have to add male and female in this data to make it multivariate then 
```

```python
sns.scatterplot( x = tips['total_bill'], y = tips['tip'],hue=tips['sex'] , style=tips['smoker']) # this is bivariate 
# now we also have to add male and female in this data to make it multivariate then 
# here using hue we can see the sex of the population who tips and the total-bill and tips 
# this is called multivariate becuase we have added 4 column in one 
```

## 2.Bar Plot (Numerical - Categorical)

```python
titanic.head()
```

```python


sns.barplot(x=titanic['Pclass'], y=titanic['Age'], hue=titanic['Pclass'], palette=['red', 'blue', 'green'])

```

```python


sns.barplot(x=titanic['Pclass'], y=titanic['Age'], hue=titanic['Sex'], palette=['red', 'blue'])

```

## 3.Box Plot (Numerical - Categorical)

```python
sns.boxplot( x= titanic['Sex'] , y = titanic['Age'] ,hue = titanic['Survived'] , palette=['red','blue'])
```

## 4.Distplot (Numerical - Categorical)

```python
sns.kdeplot(titanic[titanic['Survived'] == 0]['Age'] ) 
sns.kdeplot(titanic[titanic['Survived'] == 1]['Age'] ) 
# probabilit density function ho yo chai 
# distplot is not that used nowadays instead use kdeplot
```

## 5.HeatMap (Categorical - Categorical)

```python
titanic.head(3)
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'] , titanic['Survived']))
```

```python
(titanic.groupby('Pclass').mean(numeric_only=True)['Survived']*100).plot(kind = 'bar' )
```

```python
titanic.groupby('Sex').mean(numeric_only=True)['Survived']*100
```

## 6.ClusterMap (Categorical - Categorical)

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
# which which are closer to one another it finds that
```

## 8.LinePlot (Numerical - Numerical)

```python
flights.head()
```

```python
new = flights.groupby('year').sum(numeric_only=True).reset_index()
```

```python
sns.lineplot( x = new['year'] , y = new['passengers'])
# there is linear growth in flight between 1950 to 1960 
```

```python
sns.heatmap(flights.pivot_table(values = 'passengers' , index = 'month' ,columns = 'year'))
```

```python
sns.clustermap(flights.pivot_table(values = 'passengers' , index = 'month' ,columns = 'year'))
```
