### Missing Indicator

Suppose we have to column age and fare now if there is missing column in any of the column then we indicate and make a new column and indicate true and false if there is missing column then set it true and id there is no missing value then set it to false

```python
import numpy as np 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.impute import MissingIndicator , SimpleImputer
```

```python
df = pd.read_csv('train.csv',usecols=['Age','Fare','Survived'])
```

```python
df.head()
```

```python
X = df.drop(columns=['Survived'])
y = df['Survived']
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size = 0.2 , random_state =2)
```

```python
X_train.head()
```

```python
si = SimpleImputer()
X_train_trf = si.fit(X_train)
X_test_trf = si.fit(X_test)
X_train_trf = si.transform(X_train)
X_test_trf = si.transform(X_test)
```

```python
X_train_trf
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
clf.fit(X_train_trf, y_train)
y_pred = clf.predict(X_test_trf)
from sklearn.metrics import accuracy_score
accuracy_score (y_test , y_pred)
```

```python
mi = MissingIndicator()
mi.fit(X_train)
```

```python
X_train_missing = mi.transform(X_train)
X_train_missing
```

```python
X_test_missing = mi.transform(X_test)
X_test_missing
```

```python
X_train['Age_NA'] = X_train_missing
```
