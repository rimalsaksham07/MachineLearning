### By default aba data ma time ani date xa bhane it is on string so we cant use now we will convert it into date and time

```python
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

## Extract year

```python
date['date_year'] = date['date'].dt.year
date.head()
```

### Extract Month

```python
date['date_month'] = date['date'].dt.month
date.head()
```

```python
date['date_month'] = date['date'].dt.month_name()
date.head()
```

### Extract days


```python
date['no_days'] = date['date'].dt.day
date.head()
```

```python
date['date_dow'] = date['date'].dt.dayofweek
date.head()
```

```python

# day of week - name

date['date_dow_name'] = date['date'].dt.day_name()

date.drop(columns=['product_id','city_id','orders']).head()
```

```python

# is weekend?

date['date_is_weekend'] = np.where(date['date_dow_name'].isin(['Sunday', 'Saturday']), 1,0)

date.drop(columns=['product_id','city_id','orders']).head()
```

EXtract Qurater

```python

date['quarter'] = date['date'].dt.quarter

date.drop(columns=['product_id','city_id','orders']).head()
```

```python

date['semester'] = np.where(date['quarter'].isin([1,2]), 1, 2)

date.drop(columns=['product_id','city_id','orders']).head()
```
