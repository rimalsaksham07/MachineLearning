## OneHot Encoding

It is used for handling nominal categorical data 
for eg in the dataset it is given - good , bad , average then we use ordinal becuase in this we can give the order of the value 
but for eg in the dataset if it is given yellow , blue , red then there is no any order in this data set so we cant use ordinal data instead we use nominal encoding for eg onehot encoding in this we convert the string in the vector it will divide the color into different columns and make different columns for the values
