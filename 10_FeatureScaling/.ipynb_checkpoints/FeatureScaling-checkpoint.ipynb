## Feature Scaling
Feature scaling is a technique to standarizr the independent features present in the data in a fixed range 



## types of feature scaling
## 1) Standarization 
## 2) Normalization - min max scaler etc

## Standardization also called Z-score Normalization 

```python
# age  #salary
27     
15
33
90 
# Xi - X(mean) / sigma 
# we standarize the dataset using standard deviation 
# mean and standard deviation 
# so now we standardize the dataset there weill be new data 
# now mean will be 0 and sigma = 1 
```

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
df.head()
```

```python
df = df.iloc[:,2:]
```

```python
df
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(df.drop('Purchased' , axis=1 ) , df['Purchased'], test_size = 0.3 , random_state = 0)
X_train.shape , X_test.shape
```

## Standard Scaler

```python
from sklearn.preprocessing import StandardScaler
# class import gareko yo formula wala
scaler = StandardScaler()
scaler.fit(X_train) 
# mean calcukate garxa ani store garera rakxa mean calulate garxa ani store garxa 
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
# transform duitai train chai euta matra
```

```python
X_test_scaled
```

```python
scaler.mean_
```

```python
X_train_scaled
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled , columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_train_scaled , columns=X_test.columns)
```

```python
X_train_scaled
```

```python
X_test_scaled
```

```python
np.round(X_train.describe() , 1)
```

```python
np.round(X_train_scaled.describe() , 1)
```

## Effect Of Scaling

```python
fig , (ax1 , ax2) = plt.subplots(ncols = 2 , figsize=(12,5))
ax1.scatter(X_train['Age'], X_train['EstimatedSalary'])
ax1.set_title('Before Scaling')
ax2.scatter(X_train_scaled['Age'], X_train_scaled['EstimatedSalary'],color = 'red')
ax2.set_title("After Scaling")
plt.show()
```

```python
# distribution same hunxa tara sacle chai change hunxa 
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr_scaled = LogisticRegression()
```

```python
lr.fit(X_train , y_train)
lr_scaled.fit(X_train_scaled , y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred_scaled = lr_scaled.predict(X_test_scaled)
```

```python
y_pred_scaled
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Actual",accuracy_score(y_test , y_pred))
```

```python
print("Scaled",accuracy_score(y_test, y_pred_scaled))
```

# NORMALIZATION


##### Nomralization is a technique often applied as part of data preparation for machine learning The goal of normalization is to change the values of numeric columns in the dataset to use a common scale 

### Types of Normalization 
#### MinMax Scaling -- this is the most important one 
#### Mean Normalization 
#### Max Absolute
#### Robust Scaling 

## MinMax Scaling


```python
## for example we have weight and now we have to use normalization we use formula 
## xi = Xi - Xmin / Xmax - Xmin 
## range [0,1 ] mai hunxa output jaile 
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv('wine_data.csv', header=None , usecols=[0,1,2])
df.columns = ['Class label' , 'Alcohol' , 'Malic acid']
```

```python
df
```

```python
sns.kdeplot(df['Alcohol'])
```

```python
color_dict = {1:'red',3:'green',2:'blue'}
sns.scatterplot(x=df['Alcohol'], y=df['Malic acid'], hue=df['Class label'], palette=color_dict)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train ,X_test ,  y_train , y_test = train_test_split(df.drop('Class label',axis=1),df['Class label'],test_size = 0.3 , random_state = 0 )
```

```python
X_train.shape , X_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled )
X_test_scaled = pd.DataFrame(X_test_scaled )
X_train_scaled
```

```python
np.round(X_train.describe() , 1)
```
