```python
# url bata load garna ni milxa 
import requests
from io import StringIO
url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS x 10.14; rv:66.0 Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers)
data = StringIO(req.text)
pd.read_csv(data)
# aru same hunxa url ma gayera change garne its just a code snippet 
```

```python
#skip bad lines 
pd.read_csv('NAFLD.csv' , encoding='latin-1' )
# error_bad_lines= False kei row ma error xa bhane yo use garne error janxea
```
