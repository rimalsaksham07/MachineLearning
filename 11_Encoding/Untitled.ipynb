### Ordinal Encoding vs Label Encoding 

Ordinal Encoding ma chai edi hamro input ma kei string xa bhane we use ordinal encoding 

but if there is any string in the output classification like Yes/No bhayo bhane chai label encoding use hunxa 

```python
import numpy as np 
import pandas as pd
```

```python
df = pd.read_csv('customer.csv')
```

```python
df.sample(5)
```

```python
# gender,review , education , purchased  - cateogrical 
```

### review , education - ordinal
### purchaged - label encoder

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(df.iloc[:,2:4],df.iloc[: ,-1],test_size=0.2)
```

```python
X_train
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
# categories is going to be a list where we give order for the column in order poor < Average <Good
```

```python
oe.fit(X_train)
```

```python
X_train = oe.transform(X_train)
X_test = oe.transform(X_test)
```

```python
X_train
```

```python
oe.categories
```
