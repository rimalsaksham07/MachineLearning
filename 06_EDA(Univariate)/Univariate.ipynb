# 1.Categorical Data


## a.Countplot


```python
# it is nothing but just counting the categprical data of the given data categorical data ma chai ki barchart ki pie char use garxina
df['Pclass'].value_counts().plot(kind = 'bar')
# df['Survived'].value_counts().plot(kind='bar')
```

```python
sns.countplot(x=df['Sex'])
```

```python
df['Pclass'].value_counts().plot(kind='pie' ,autopct='%2f')
```

# Numerical Data
## a.Histogram 

```python
# now if we have numercial data column then what are the information we can gathered from numerical data
# calcukating mean and soo on 
# we can first convert the numercial data into categorical data (0-10) , (10-20) esari category ma rakhne
# numerical data xa bhane it is best to use histogram 

import matplotlib.pyplot as plt
plt.hist(df['Age'] ,bins=50)

# to understand distribution of data in the speicific column in univariate
```

## b.Displot

```python
sns.distplot(df['Age'])

# in distplot we find the probability of having that in speicific group gives us a probability 
# PDF (probability distribution function)
```

## c.Boxplot

```python
# inter quaritle range haru wala 
# 5 number summary 
# used for detecting the outliers
sns.boxplot(x = df['Fare'] )
```
